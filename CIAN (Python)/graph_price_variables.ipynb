```python
import pandas as pd
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Импорт данных
data_second = pd.read_csv(r'C:\Data_P\GitHub\data_second_january_2021.csv')
```

```python
data_second
```

### 1) Зависимость между логарифмом цен и некоторыми числовыми переменными

```python
# Вычисление логарифма цены
data_second['log_price'] = float('nan')
for i in range(0, len(data_second)):
    data_second.loc[i, 'log_price'] =  math.log(data_second.loc[i, 'price'])
```

```python
# Общая площадь
sns.set_theme(font_scale = 1)
sns.mpl.rc('figure', figsize = (30,20))
sns.relplot(x="square_total", y="log_price",  palette='colorblind',
            hue="rooms", data=data_second)

# Площадь кухни
sns.set_theme(font_scale = 1)
sns.mpl.rc('figure', figsize = (30,20))
sns.relplot(x="square_kitchen", y="log_price",  palette='colorblind',
            hue="rooms", data=data_second)

# Жилая площадь
sns.set_theme(font_scale = 1)
sns.mpl.rc('figure', figsize = (30,20))
sns.relplot(x="square_living", y="log_price",  palette='colorblind',
            hue="rooms", data=data_second)

# Высота потолков
sns.set_theme(font_scale = 1)
sns.mpl.rc('figure', figsize = (30,20))
sns.relplot(x="ceiling", y="log_price",  palette='colorblind',
            hue="rooms", data=data_second)

# Год постройки
sns.set_theme(font_scale = 1)
sns.mpl.rc('figure', figsize = (30,20))
sns.relplot(x="built_year", y="log_price",  palette='colorblind',
            hue="rooms", data=data_second)

# Этаж квартиры
sns.set_theme(font_scale = 1)
sns.mpl.rc('figure', figsize = (30,20))
sns.relplot(x="floor_flat", y="log_price",  palette='colorblind',
            hue="rooms", data=data_second)

# Количество этажей в доме
sns.set_theme(font_scale = 1)
sns.mpl.rc('figure', figsize = (30,20))
sns.relplot(x="floor_house", y="log_price",  palette='colorblind',
            hue="rooms", data=data_second)
```
