## Независимые переменные

### 1) Распределение независимых переменных

```python
list_variable = ['rooms', 'lat', 'lng', 'square_total', 'square_living',
       'square_kitchen', 'type_apartment', 'ceiling', 'renovation', 'view',
       'type_house', 'floor_type', 'built_year',
       'playground', 'sportsground', 'foundation_type', 'electrical_type', 'cold_water_type', 'sewerage_type',
       'gas_type', 'ventilation_type', 'firefighting_type', 'drainage_type',
       'floor_flat']


fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4,figsize=(20,6))
ax_num = (ax1, ax2, ax3, ax4)
for i in range(0, len(ax_num[0:4])):
    ax_num[i].tick_params(rotation=45, axis='x')
    sns.set(font_scale = 1)
    sns.mpl.rc('figure', figsize = (20,6))
    sns.histplot(data_second[list_variable[i]], ax=ax_num[i]).set(ylabel='', xlabel='', title = 'variable = ' + list_variable[i])

ax_num = (ax1, ax2, ax3, ax4) 
fig, ax_num = plt.subplots(1, 4,figsize=(20,6))
for i in range(0, len(ax_num)):
    ax_num[i].tick_params(rotation=45, axis='x')
    sns.set(font_scale = 1)
    sns.mpl.rc('figure', figsize = (20,6))
    plt.xticks(rotation=45)
    sns.histplot(data_second[list_variable[4+i]], ax=ax_num[i]).set(ylabel='', xlabel='', title = 'variable = ' + list_variable[4+i])
    
ax_num = (ax1, ax2, ax3, ax4) 
fig, ax_num = plt.subplots(1, 4,figsize=(20,6))
for i in range(0, len(ax_num)):
    ax_num[i].tick_params(rotation=45, axis='x')
    sns.set(font_scale = 1)
    sns.mpl.rc('figure', figsize = (20,6))
    plt.xticks(rotation=45)
    sns.histplot(data_second[list_variable[8+i]], ax=ax_num[i]).set(ylabel='', xlabel='', title = 'variable = ' + list_variable[8+i])

ax_num = (ax1, ax2, ax3, ax4) 
fig, ax_num = plt.subplots(1, 4,figsize=(20,6))
for i in range(0, len(ax_num)):
    ax_num[i].tick_params(rotation=45, axis='x')
    sns.set(font_scale = 1)
    sns.mpl.rc('figure', figsize = (20,6))
    plt.xticks(rotation=45)
    sns.histplot(data_second[list_variable[12+i]], ax=ax_num[i]).set(ylabel='', xlabel='', title = 'variable = ' + list_variable[12+i])

ax_num = (ax1, ax2, ax3, ax4) 
fig, ax_num = plt.subplots(1, 4,figsize=(20,6))
for i in range(0, len(ax_num)):
    ax_num[i].tick_params(rotation=45, axis='x')
    sns.set(font_scale = 1)
    sns.mpl.rc('figure', figsize = (20,6))
    plt.xticks(rotation=45)
    sns.histplot(data_second[list_variable[16+i]], ax=ax_num[i]).set(ylabel='', xlabel='', title = 'variable = ' + list_variable[16+i])

ax_num = (ax1, ax2, ax3, ax4) 
fig, ax_num = plt.subplots(1, 4,figsize=(20,6))
for i in range(0, len(ax_num)):
    ax_num[i].tick_params(rotation=45, axis='x')
    sns.set(font_scale = 1)
    sns.mpl.rc('figure', figsize = (20,6))
    plt.xticks(rotation=45)
    sns.histplot(data_second[list_variable[20+i]], ax=ax_num[i]).set(ylabel='', xlabel='', title = 'variable = ' + list_variable[20+i])
```

### 2) Корреляционная матрица числовых переменных (в т.ч. географических координат)

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(data_second.corr(), xticklabels=data_second.corr().columns, yticklabels=data_second.corr().columns, cmap='RdYlGn', center=0, annot=False)

plt.title('Корреляционная матрица числовых переменных', fontsize=26)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
