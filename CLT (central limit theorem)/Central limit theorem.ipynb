```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Для данного задания было взято экспоненциальное распределение. В данном случае, lambda = 1, а значит, функция плотности распределения имеет вид: f(x) = exp(-x) для x >=0 и f(x) = 0, если x < 0. 

```python
# Экспоненциальное распределение (exp(-x)). По умолчанию sts.exp() строитс экспоненциальное распределение с параметром lambda = 1.
exp_rv = sts.expon()
# Генерируем выборку размера 1000.
sample = exp_rv.rvs(size = 1000)
```

```python
# Генерируем выборку размера 1000.
x = np.linspace(0,10,1000)
# Строим гистограмму, подписываем оси и т. д.
plt.hist(sample, density=True, label='histogram', bins=60)
plt.ylabel('f(x)')
plt.xlabel('$x$')
# Рисуем теоретическую плотность распределения.
pdf = exp_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
```

Поскольку в данном случае рассмотрено экспоненциальное распределение с параметром lambda = 1, то согласно известным формулам (https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D1%81%D0%BF%D0%BE%D0%BD%D0%B5%D0%BD%D1%86%D0%B8%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5):

Математическое ожидание (E(x)) = 1;

Дисперсия (D(x)) = 1.

Следовательно, распределение нашего выборочного среднего будет: N(1;$\frac{1}{\sqrt{n}})$. 

a) Для n = 5: N(1, $\frac{1}{\sqrt{5}})$

b) Для n = 10: N(1, $\frac{1}{\sqrt{10}})$)

c) Для n = 50: N(1, $\frac{1}{\sqrt{50}})$)

Сгенерируем 1000 выборок объёма n (были взяты значения 5, 10, 50) и построим гистограммы распределений выборочных средних.


## Случай n = 5

```python
# Вычислили выборочные среднии по 1000 выборкам, состоящих из 5 элементов.
means = []
for i in range(1000):
    means.append(exp_rv.rvs(5).mean())
# Гистограмма
plt.hist(means, bins = 60, density=True, range=[0, 4], label='means histogram')
plt.ylabel('$f(x)$')
plt.xlabel('mean $x$')
## Вычисляем и рисуем теоретическую плотность распределения.
norm_rv = sts.norm(1, np.sqrt(1/5))
x = np.linspace(0, 4, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='norm theoretical pdf')
plt.legend()
```

## Случай n = 10

```python
# Вычислили выборочные среднии по 1000 выборкам, состоящих из 10 элементов.
means = []
for i in range(1000):
    means.append(exp_rv.rvs(10).mean())
# Гистограмма
plt.hist(means, bins = 60, density=True, range=[0, 4], label='means histogram')
plt.ylabel('$f(x)$')
plt.xlabel('mean $x$')
## Вычисляем и рисуем теоретическую плотность распределения.
norm_rv = sts.norm(1, np.sqrt(1/10))
x = np.linspace(0, 4, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='norm theoretical pdf')
plt.legend()
```

## Случай n = 50

```python
# Вычислили выборочные среднии по 1000 выборкам, состоящих из 50 элементов.
means = []
for i in range(1000):
    means.append(exp_rv.rvs(50).mean())
# Гистограмма
plt.hist(means, bins = 60, density=True, range=[0, 4], label='means histogram')
plt.ylabel('$f(x)$')
plt.xlabel('mean $x$')
## Вычисляем и рисуем теоретическую плотность распределения.
norm_rv = sts.norm(1, np.sqrt(1/50))
x = np.linspace(0, 4, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='norm theoretical pdf')
plt.legend()
```

По мере роста n, гистограмма все больше начинает походить по форме на нормальное распределение. При этом, распределение средних значений начинает сходиться к 1, а разброс становится меньше.
